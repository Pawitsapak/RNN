```
import dill
with open('thai-ner/model/1.2/datatrain.data', 'rb') as file:
 datatofile = dill.load(file)
```

```
max(len(sen) for sen in datatofile)
```

```
word_ner=[]
for i in datatofile:
    t1=[]
    t2=[]
    for j in i:
        t1.append(j[0])
        t2.append(j[2])
    word_ner.append((t1,t2))
```

```
word_list=[]
pos_list=[]

for sent,tags in word_ner:
    for word in sent:
        word_list.append(word)
    for tag in tags:
        pos_list.append(tag)

#Word to Index

all_words = sorted(set(word_list))
all_pos = sorted(set(pos_list))
#del word_list, pos_list
word_to_ix = dict((c, i+2) for i, c in enumerate(all_words)) #convert word to index 
tag_to_ix = dict((c, i+1) for i, c in enumerate(all_pos)) #convert pos to index
```

```
word_to_ix["UNK"] = 1
word_to_ix["PAD"] = 0
tag_to_ix["PAD"] = 0
```

```
ix_to_word = {i: w for w, i in word_to_ix.items()}
ix_to_tag = {i: w for w, i in tag_to_ix.items()}
```

```
from keras.preprocessing.sequence import pad_sequences
X_word = [[word_to_ix[w[0]] for w in s] for s in datatofile] #เปลี่ยนคำเป็นเลข
```

```
max_len = 300
max_len_char = 90
```

```
X_word = pad_sequences(maxlen=max_len, sequences=X_word, value=word_to_ix["PAD"], padding='post', truncating='post') #หลังจากทำคำเป็นเลข
```

```
words = list(set(word_list))
```

```
chars = set([w_i for w in words for w_i in w])
n_chars = len(chars)
print(n_chars)
```

```
char_to_ix = {c: i + 2 for i, c in enumerate(chars)}
char_to_ix["UNK"] = 1
char_to_ix["PAD"] = 0
```

```
import numpy as np

X_char = []
for sentence in datatofile:
    sent_seq = []
    for i in range(max_len):
        word_seq = []
        for j in range(max_len_char):
            try:
                word_seq.append(char_to_ix.get(sentence[i][0][j]))
            except:
                word_seq.append(char_to_ix.get("PAD"))
        sent_seq.append(word_seq)
    X_char.append(np.array(sent_seq))
```

```
y = [[tag_to_ix[w[2]] for w in s] for s in datatofile]
y = pad_sequences(maxlen=max_len, sequences=y, value=tag_to_ix["PAD"], padding='post', truncating='post')
```

```
from sklearn.model_selection import train_test_split
X_word_tr, X_word_te, y_tr, y_te = train_test_split(X_word, y, test_size=0.2, random_state=2018)
X_char_tr, X_char_te, _, _ = train_test_split(X_char, y, test_size=0.2, random_state=2018)
```

```
from keras.models import Model, Input
from keras.layers import LSTM, Embedding, Dense, TimeDistributed, Dropout, Conv1D, Bidirectional
from keras.layers import Bidirectional, concatenate, SpatialDropout1D, GlobalMaxPooling1D
from keras_contrib.layers import CRF
from keras_contrib.losses import crf_loss
from keras_contrib.metrics import crf_viterbi_accuracy
```

```
tags = list(set(pos_list))
n_tags = len(tags)
n_words = len(words)
```

```
# input and embedding for words
word_in = Input(shape=(max_len,))
emb_word = Embedding(input_dim=n_words + 2, output_dim=20,
                     input_length=max_len, mask_zero=False)(word_in)

# input and embeddings for characters
char_in = Input(shape=(max_len, max_len_char,))
emb_char = TimeDistributed(Embedding(input_dim=n_chars + 2, output_dim=10,
                           input_length=max_len_char, mask_zero=False))(char_in)
# character LSTM to get word encodings by characters
char_enc = TimeDistributed(LSTM(units=20, return_sequences=False,
                                recurrent_dropout=0.5))(emb_char)

# main LSTM
x = concatenate([emb_word, char_enc])
x = SpatialDropout1D(0.3)(x)
main_lstm = Bidirectional(LSTM(units=50, return_sequences=True,
                               recurrent_dropout=0.6))(x)
out = TimeDistributed(Dense(n_tags + 1, activation="relu"))(main_lstm)
crf = CRF(n_tags+1)  # CRF layer
out = crf(out)  # output
model = Model([word_in, char_in], out)
```

```
model.compile(optimizer="adam", loss=crf_loss, metrics=[crf_viterbi_accuracy])
```

```
model.summary()
```

```
from keras.utils import to_categorical
y_tr = to_categorical(np.array(y_tr), num_classes=len(tag_to_ix))
```

```
history = model.fit([X_word_tr,
                     np.array(X_char_tr)],
                    y_tr,
                    batch_size=32, epochs=15, validation_split=0.1, verbose=1)
```

```
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelBinarizer, LabelEncoder
from itertools import chain

def bio_classification_report(y_true, y_pred):
    """
    Classification report for a list of BIO-encoded sequences.
    It computes token-level metrics and discards "O" labels.
    
    Note that it requires scikit-learn 0.15+ (or a version from github master)
    to calculate averages properly!
    """
    lb = LabelEncoder()
    flattened_y_true = list(chain.from_iterable(y_true))
    flattened_y_pred = list(chain.from_iterable(y_pred))
    y_true_combined = lb.fit_transform(flattened_y_true)
    y_pred_combined = lb.transform(flattened_y_pred)

    tagset = set(lb.classes_) - {'O', 'PAD','I-EMAIL','B-EMAIL','B-ZIP','B-DATE','I-DATE','B-LAW','I-LAW','B-LEN','I-LEN','B-MONEY','I-MONEY','B-PERCENT','I-PERCENT','B-TIME','I-TIME','B-PHONE','I-PHONE','B-URL','I-URL'} #{'O', 'ENDPAD'}
    tagset = sorted(tagset, key=lambda tag: tag.split('-', 1)[::-1])
    class_indices = {cls: idx for idx, cls in enumerate(lb.classes_)}
    
    accuracy = np.mean([x==y for x, y in zip(flattened_y_true, flattened_y_pred) if x != 'ENDPAD'])
    print(np.round(accuracy, 2))

    print (classification_report(
        y_true_combined,
        y_pred_combined,
        labels = [class_indices[cls] for cls in tagset],
        target_names = tagset)
    ) 
```

```
def predict(input): #input > list of lists (1 corpus)
    X_char = []
    X_word = []
    for s in input:
        pp = []
        for w in s:
            pp.append(word_to_ix.get(w, word_to_ix["UNK"]))
        X_word.append(pp)
  
    
    X_word = pad_sequences(maxlen=max_len, sequences=X_word, value=word_to_ix["PAD"], padding='post', truncating='post') #หลังจากทำคำเป็นเลข
    print(X_word.shape)
    
    for sentence in input:
        sent_seq = []
        for k in range(max_len):
            word_seq = []
            for j in range(max_len_char):
                try:
                    word_seq.append(char_to_ix.get(sentence[k][j], char_to_ix['UNK']))
                except:
                    word_seq.append(char_to_ix.get("PAD"))
            sent_seq.append(word_seq)
        X_char.append(np.array(sent_seq))
    print(np.array(X_char).shape)
    print(len(X_char))

    result = []
    y_pre = model.predict([X_word,np.array(X_char)])
    for one in y_pre:
        p = np.argmax(one, axis=-1)
        for code in p:
            if ix_to_tag[code] != 'PAD':
                result.append(ix_to_tag[code])

    
    return result

import re
from pythainlp import word_tokenize
from glob import glob 

train_filenames = glob('thai_ner_dev_set_no_answers/*.txt')

def file_to_xy(train_filenames):
    word_list = []
    
    for filename in train_filenames:
        file = []
        with open(filename, encoding="utf8") as f:
            
            
            for line in re.split(r'\n\s?\n+', f.read()):
                
                rew = []
                line = line.strip()
                phrases = word_tokenize(line, engine="newmm" )
                for i in phrases:
                    if i != '\n':              
                        rew.append(i)
                if len(rew) != 0:
                    file.append(rew)
                    
            word_list.append(file)

                            
    
    return word_list

def combine_token(input): #list of list of listของย่อหน้า
    a = []
    for doc in input:
        b = []
        c = 0
        for sen in doc:
            if c < 1:
                b += sen
                c += 1
    
        a.append(b)
    return a
```

```
def tokenize(train_filenames):
    word_list = []
    
    for filename in train_filenames:
        file = []
        with open(filename, encoding="utf8") as f:
            rew = []
            
            for line in re.split(r'\n\s?\n+', f.read()):
                line = line.strip()
                phrases = word_tokenize(line, engine="newmm" )
                for i in phrases:
                    if i != '\n':
                                            
                        rew.append(i)
                
                    
            word_list.append(rew)

                            
    
    return word_list
```

```
def extract_word2(WordAndTag,file): #input = tuple >> (list of word, list of tag) file = str of filename after glob
    ans = {}
    
    #adding filename dictionary to the answer
    for file_id, filename in enumerate(train_filenames):
        namey = filename.split('/')[1]
        ans[namey] = {"place":[], "org":[], "pers":[] }
        
        #for loop list of tag in the current doc
        for tag_id, tag in enumerate(WordAndTag[1][file_id]):
            name = []
            #B and I checker
            if tag == 'B-PERSON' or tag == 'B-ORGANIZATION' or tag == 'B-LOCATION':  
                name.append(WordAndTag[0][file_id][tag_id])
                counter = 1
                try:
                    while 'I' in WordAndTag[1][file_id][tag_id + counter]:
                        name.append(WordAndTag[0][file_id][tag_id + counter])
                        counter += 1
                    else:
                        full_name = ''.join(name)
                        category = ''
                        if tag == 'B-PERSON':
                            category = 'pers'
                        elif tag == 'B-ORGANIZATION':
                            category = 'org'
                        elif tag == 'B-LOCATION':
                            category = 'place'
                        if full_name not in ans[namey][category]:
                            ans[namey][category].append(full_name)
                except:
                    full_name = ''.join(name)
                    category = ''
                    if tag == 'B-PERSON':
                        category = 'pers'
                    elif tag == 'B-ORGANIZATION':
                        category = 'org'
                    elif tag == 'B-LOCATION':
                        category = 'place'
                    if full_name not in ans[namey][category]:
                        ans[namey][category].append(full_name)
    return ans
```
